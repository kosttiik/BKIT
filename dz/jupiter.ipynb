```python
def makeURL(cnt):
    url = 'http://127.0.0.1:5000/'
    res = url + str(cnt)
    return res

def get_data(cnt):
    url = makeURL(cnt)
    r = requests.get(url)
    return r.json()
cntLst = [0, 5, 10, 12, 15, 20]
for elem in cntLst:
    print(f'{elem} первых чисел последовательности Фибоначчи: {get_data(elem)}')
```

```python
y_12 = get_data(12)
x_12 = list(range(1, len(y_12)+1))
fig = plt.figure(figsize = (7, 5))
plt.bar(x_12, y_12)
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title(f'Первые {len(y_12)} чисел последовательности Фибоначчи')
plt.show()
```
